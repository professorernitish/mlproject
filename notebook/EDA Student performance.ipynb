```python
## Define numerical and categorical columns
numeric_feature = [feature for feature in ireadCopy.columns if ireadCopy[feature].dtypes != 'O']
categorical_feature =[feature for feature in ireadCopy.columns if ireadCopy[feature].dtypes =='O']


## print columns
print('We have {} numerica feature :{} '.format(len(numeric_feature),numeric_feature))
print('\n We have {} categorical feature :{} '.format(len(categorical_feature),categorical_feature))
```

```python
## Adding columns for Total Score and Average Score
ireadCopy['total_score'] = ireadCopy['math_score']+ireadCopy['reading_score']+ireadCopy['writing_score']
ireadCopy['average'] = ireadCopy['total_score']/3
ireadCopy.head()
```

```python
reading_full = ireadCopy[ireadCopy['reading_score']==100]['average'].count()
writing_full = ireadCopy[ireadCopy['writing_score']==100]['average'].count()
math_full = ireadCopy[ireadCopy['math_score']==100]['average'].count()


print(f'Number of students with full marks in Reading :{reading_full}')
print(f'Number of students with full marks in Writing :{writing_full}')
print(f'Number of students with full marks in Math :{math_full}')
```

```python
reading_less_20 = ireadCopy[ireadCopy['reading_score']<=20]['average'].count()
writing_less_20 = ireadCopy[ireadCopy['writing_score']<=20]['average'].count()
math_less_20 = ireadCopy[ireadCopy['math_score']<=20]['average'].count()

print(f'Number of students with less than 20  marks in Reading :{reading_less_20}')
print(f'Number of students with less than 20  marks Writing :{writing_less_20}')
print(f'Number of students with less than 20  marks :{math_less_20}')
```
